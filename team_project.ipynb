From the graphs amount of reviews per sentiment and sentiment analysis we could assume that most of the people (62.1%) use the website to leave negative reviews. However, the percentage could be higher as we have found some limitations on the method used. For example some reviews that have been classified as positive contain sarcasm and if a review has several neutral words polarity compound was shifted and the review could be classified as a false positive or as a false neutral review.
This does not neccessarily accurately reflect the quality of BT broadband service, as most satisfied customers tend to not leave a review, hence the reviews of angry, unhappy customers are overrepresented in our sample.

Using collocations we saw that the top 5 two words used in the reviews were:  
'customer service' with 1066 appeareances 
'call back' with  183 appeareances 
'customer services' with  173 appeareances 
'phone line' with 165 appeareances 
'call centre'  with 156 appeareances 

From this we can infer that one of the most important features of broadband products for BT customers that use the website https://www.broadband.co.uk/broadband/providers/bt/reviews/ is the customer service 

Most of the negative reviews comes from the biggest cities of UK (London, Manchester)
Most of the negative reviews come from some of the biggest cities of the UK. We believe BT should also focus on improving the services in these areas as people from these cities will feel more compelled to complain or leave a review about their experience.

Using collocations we were able to see the top 5 most frequent word combinations (two words and three words) This allowed us to understand the nature of the reviews so we could infer that customer service is one of the most important features of broadband products for BT customers that use the website https://www.broadband.co.uk/broadband/providers/bt/reviews/ to leave reviews as the two-words combination more used were 'customer service' used 1066 times and 'customer services' used 173 times. Words like 'worst', 'terrible' and 'poor' were used to describe the customer service. 

From these results we believe BT should focus on improving the customer service as it seems to be the feature that people complain the most about.  ~
